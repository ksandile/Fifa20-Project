# Load the required Libraries to answer the following questions

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Then we read our csv file

# Now we will want to check which Nation has more players in the dataframe.

```python
fifa['nationality'].value_counts()
```

# As you can see that most of the players are from England, with 1667.

# And if we were to find a glance of first 10 countries, we gona use [0:10] method.

```python
fifa['nationality'].value_counts()[0:10]
```

# Now i want to make a bar plot for these top 5 countries, which these players belong too.

# Firstly we will check keys(index) of our dataset.

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="g")
plt.show()
```

# Now we will check the players wage salary in euros.

```python
player_salary = fifa[['short_name', 'wage_eur']]
```

```python
player_salary.head()
```

# Then you can see that Messi has the most wage, with 56500 euros.

# Now we want to see which is the most paid player?

```python
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

# You can see that Messi is the most paid player in 2022

# Now we will make a bar graph for these top 5 most paid players.

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color="r")
plt.show()
```
